```python
import pandas as pd
df = pd.read_csv('Tabaquismo.txt', nrows=7)
```

```python
cols = ['ID', 'EDAD', 'SEXO','FUMA','TIEMPO','FRECUENCIA_DIA','ENFERMEDAD',]
X = df[cols]
```

```python
X
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer()
Datos_faltantes = imputer.fit_transform(X)
```

```python
Datos_completos = pd.DataFrame(Datos_faltantes, columns=X.columns, index=X.index)
```

```python
print(Datos_completos)
```
